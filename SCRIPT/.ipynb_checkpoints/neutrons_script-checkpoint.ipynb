```python
#CONSTANTS 

#atmospheric corrections
NINC_REF = 150
RH_REF = 0
ALPHA = 0.0054
BETA = 0.0076
P_REF = 973.4

#conversion to soil moisture
A0 = 0.0808
A1 = 0.372
A2 = 0.115
N0 = 1577
BD = 1.5
theta_offset = 0

###################################################

import pandas as pd
import csv
import numpy
import matplotlib.pyplot as plt

#fill holes in datasets
#datasets used are: raw data from finapp, elaborated data from finapp, incoming data from https://www.nmdb.eu/nest/search.php, hourly ERG5 data (prec, tavg, RH)

#plt.params

print("Starting up...")
print("You'll need three types of file in the same folder as this script: raw.csv, incoming.csv, ERG5.csv, finapp.csv. \n")
#input("Check that they're there and press any key to continue. \n\n")
print("Check that they're there and press any key to continue. \n\n")
print("Reading and tidying data...", end="")
df = pd.read_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/raw.csv', sep=';', parse_dates=['#Datetime'])
start = df['#Datetime'].min()
end = df['#Datetime'].max()
df = df.drop_duplicates(subset='#Datetime')
new_index = pd.date_range(start=start, end=end, freq='H')
df = df.set_index('#Datetime').reindex(new_index)
df.to_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/ELABORATIONS/raw_corretto.csv')

df = pd.read_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/incoming.csv', sep=';', parse_dates=["start_date_time"])
start = df['start_date_time'].min()
end = df['start_date_time'].max()
df = df.drop_duplicates(subset='start_date_time')
new_index = pd.date_range(start=start, end=end, freq='H')
df = df.set_index('start_date_time').reindex(new_index)
df.to_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/ELABORATIONS/incoming_corretto.csv')

df = pd.read_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/ERG5.csv', parse_dates=["PragaTime"])
start = df['PragaTime'].min()
end = df['PragaTime'].max()
df = df.drop_duplicates(subset='PragaTime')
new_index = pd.date_range(start=start, end=end, freq='H')
df = df.set_index('PragaTime').reindex(new_index)
df.to_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/ELABORATIONS/ERG5_corretto.csv')
print(" Done!\n")

#load datasets and merge them

print("Preparing dataset...", end="")
df1 = pd.read_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/ELABORATIONS/raw_corretto.csv', parse_dates=[0])
df2 = pd.read_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/ELABORATIONS/incoming_corretto.csv', parse_dates=[0])
df3 = pd.read_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/ELABORATIONS/ERG5_corretto.csv', parse_dates= [0])

merged = pd.merge(df1, df2, how = 'inner', left_index = True, right_index = True)
data = pd.merge(merged, df3, how = 'inner', left_index = True, right_index = True)
data.drop(data.columns[[3,4,5,6,7,8,16,17,18,19,20,21]], axis =1, inplace=True) #drop unused columns
data.drop(data.columns[[4,6]], axis = 1, inplace = True) #drop two other datetime columns
data.rename(columns={"Unnamed: 0_x":"datetime","pressure(hPa)":"PRESS(hPa)","RCORR_E":"incoming"}, inplace=True)

data.to_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/RESULTS/whole_data.csv')

finapp = pd.read_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/finapp.csv', sep=';', parse_dates=['#Datetime'])
start = finapp['#Datetime'].min()
end = finapp['#Datetime'].max()
finapp = finapp.drop_duplicates(subset='#Datetime')
new_index = pd.date_range(start=start, end=end, freq='H')
finapp = finapp.set_index('#Datetime').reindex(new_index)
finapp['soil_moisture'] = finapp['soil_moisture']*BD

print(" Done! \nYou can find the clean data in the same folder as this script.\n")

#ATMOSPHERIC CORRECTIONS
print("Calculating atmospheric corrections and corrected neutron count...", end ="")
RH_REF = data["RHAVG"].mean()
data["incoming"]=data['incoming'].rolling(6, center=True).mean()
data["incoming_correction"] = NINC_REF/data["incoming"]
data["press_correction"] = numpy.exp(BETA*(data["PRESS(hPa)"]-P_REF))
#data["RH_correction"] = (1+ALPHA*(data["RHAVG"]/100-RH_REF/100))
data['RH_correction'] = 1
data["corrected_neutrons"] = data["neutrons"]*data["incoming_correction"]*data["press_correction"]*data["RH_correction"]
data['movingAvg_neutrons'] = data['corrected_neutrons'].rolling(12, center=True).mean()
data['soil_moisture'] = (A0/((data["movingAvg_neutrons"]/N0)-A1)-A2-theta_offset)*BD
print(" Done!")

#for later use
corrections = data.copy()


#media giornaliera 
print("Calculating soil moisture and elaborathing graphs...")
data = data.set_index(pd.DatetimeIndex(data['datetime']))
dailyPrec = data.copy()
dailyPrec.drop(dailyPrec.columns[[1,2,3,4,5,7,8,9,10,11,12,13]], axis =1, inplace=True)
dailyData = data.resample('D').mean()
del(dailyPrec["datetime"])
dailyPrec = dailyPrec.resample("D").sum()
dailyData.to_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/RESULTS/dailySoilMoisture.csv')
dailyFinapp = finapp.resample("D").mean()
dailyFinapp.drop(dailyFinapp.columns[[1,2,3,4,5,6]], axis =1, inplace=True)






#grafici giornalieri di SM
dailyData["datetime"] =pd.to_datetime(dailyData["datetime"])
fig, ax = plt.subplots(figsize=(22, 5))
ax2 = ax.twinx()
x = dailyData.index
ax.plot(x, dailyData['soil_moisture'], label='soil moisture', color ="red")
ax2.bar(x, dailyPrec['PREC'], width=2, label='prec',edgecolor = 'black')

ax.set_ylabel("soil moisture",fontsize=12)
ax2.set_ylabel("precipitation (mm)",fontsize=12)
ax.axhline(y=0.42, ls='--', label="saturation", color = 'blue')
ax.axhline(y=0.36, ls='--', label="field capacity", color = 'skyblue')
ax.axhline(y=0.19, ls='--', label="wilting point", color = 'gold')
ax.set_title("DAILY - CRNS soil moisture vs prec",fontsize=18)
ax.legend(fontsize=13,ncol = 3, loc='upper right')
ax2.legend(fontsize=13, ncol = 3, loc="upper left")
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
plt.tight_layout()
plt.show()
fig.savefig('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/RESULTS/dailySM.png')

#precipitazione semi-mensile e soil moisture semi mensile
semiMonthlyPrec = dailyPrec.resample("SM").sum()
semiMonthlyData = data.resample("SM").mean()
semiMonthlyData.to_csv('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/RESULTS/semiMonthlySoilMoisture.csv')

"""semiMonthlyData["datetime"] =pd.to_datetime(semiMonthlyData["datetime"])
fig, ax = plt.subplots(figsize=(22, 5))
ax2 = ax.twinx()
x = semiMonthlyData.index
ax2.bar(x, semiMonthlyPrec['PREC'], width=10, label='prec',edgecolor = 'black')
ax.plot(x, semiMonthlyData['soil_moisture'], label='soil moisture', color ="red")
ax.axhline(y=0.42, ls='--', label="saturation", color = 'blue')
ax.axhline(y=0.36, ls='--', label="field capacity", color = 'skyblue')
ax.axhline(y=0.19, ls='--', label="wilting point", color = 'gold')

ax.set_ylabel("soil moisture",fontsize=12)
ax2.set_ylabel("precipitation (mm)",fontsize=12)
ax.set_title("SEMIMONTHLY - CRNS soil moisture vs prec",fontsize=18)
ax.legend(fontsize=13,ncol = 3, loc='upper right')
ax2.legend(fontsize=13, ncol = 3, loc="upper left")
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
plt.tight_layout()
plt.show()
fig.savefig('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/RESULTS/semiMonthlySM.png')"""

#confronto con elaborazione finapp
semiMonthlyFinapp = finapp.resample("SM").mean()
semiMonthlyFinapp.drop(semiMonthlyFinapp.columns[[1,2,3,4,5,6]], axis =1, inplace=True)
semiMonthlyFinapp.rename(columns={"soil_moisture":"finapp_soil_moisture"}, inplace=True)
semiMonthlyPrec.rename(columns={"PREC":"cumulativePREC"}, inplace=True)
semiMonthlyData = pd.merge(semiMonthlyData, semiMonthlyFinapp, how = 'inner', left_index = True, right_index = True)
semiMonthlyData = pd.merge(semiMonthlyData, semiMonthlyPrec, how = 'inner', left_index = True, right_index = True)

fig, ax = plt.subplots(figsize=(22, 5))
ax2 = ax.twinx()
x = semiMonthlyData.index
ax2.bar(x, semiMonthlyData['cumulativePREC'], width=7, label='prec',edgecolor = 'black')
ax.plot(x, semiMonthlyData['soil_moisture'], label='soil moisture', color ="red")
ax.plot(x, semiMonthlyData['finapp_soil_moisture'], label="Finapp soil Moisture", color='green')
ax.axhline(y=0.42, ls='--', label="saturation", color = 'blue')
ax.axhline(y=0.36, ls='--', label="field capacity", color = 'skyblue')
ax.axhline(y=0.19, ls='--', label="wilting point", color = 'gold')

ax.set_ylabel("soil moisture",fontsize=12)
ax2.set_ylabel("precipitation (mm)",fontsize=12)
ax.set_title("SEMIMONTHLY - CRNS soil moisture vs Finapp data vs prec",fontsize=18)
ax.legend(fontsize=13,ncol = 3, loc='upper right')
ax2.legend(fontsize=13, ncol = 3, loc="upper left")
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
plt.tight_layout()
plt.show()
fig.savefig('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/RESULTS/semiMonthlySM.png')






#alluvioni 
finapp.drop(finapp.columns[[1,2,3,4,5,6]], axis =1, inplace=True)
finapp.rename(columns={"soil_moisture":"finapp_soil_moisture"}, inplace=True)
data = pd.merge(data, finapp, how = 'inner', left_index = True, right_index = True)

fig, ax = plt.subplots(figsize=(22, 5))
ax2 = ax.twinx()
x = data.index

ax.plot(x, data['soil_moisture'], label='soil moisture', color ="red")
ax.plot(x, data['finapp_soil_moisture'], label='finapp soil moisture', color ="green")
ax2.bar(x, data['PREC'], width=0.1, label='prec',edgecolor = 'black')


ax.set_ylabel("soil moisture",fontsize=12)
ax2.set_ylabel("precipitation (mm)",fontsize=12)
ax.axhline(y=0.42, ls='--', label="saturation", color = 'blue')
ax.axhline(y=0.36, ls='--', label="field capacity", color = 'skyblue')
ax.axhline(y=0.19, ls='--', label="wilting point", color = 'gold')
ax.set_xlim(pd.Timestamp('2023-05-01'), pd.Timestamp('2023-06-01'))
ax.set_ylim(0,0.6)
ax.set_title("Maggio 2023 - soil moisture vs Finapp soil moisture vs prec",fontsize=18)
ax.legend(fontsize=13,ncol = 3, loc='upper left')
ax2.legend(fontsize=13, ncol = 3, loc="upper right")
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
plt.tight_layout()
plt.show()
fig.savefig('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/RESULTS/maggio2023.png')

#ottobre
fig, ax = plt.subplots(figsize=(22, 5))
ax2 = ax.twinx()
x = data.index

ax.plot(x, data['soil_moisture'], label='soil moisture', color ="red")
ax.plot(x, data['finapp_soil_moisture'], label='finapp soil moisture', color ="green")
ax2.bar(x, data['PREC'], width=0.1, label='prec',edgecolor = 'black')


ax.set_ylabel("soil moisture",fontsize=12)
ax2.set_ylabel("precipitation (mm)",fontsize=12)
ax.axhline(y=0.42, ls='--', label="saturation", color = 'blue')
ax.axhline(y=0.36, ls='--', label="field capacity", color = 'skyblue')
ax.axhline(y=0.19, ls='--', label="wilting point", color = 'gold')
ax.set_xlim(pd.Timestamp('2024-09-01'), pd.Timestamp('2024-10-31'))
ax.set_ylim(0,0.6)
ax.set_title("Ottobre 2024 - soil moisture vs Finapp soil moisture vs prec",fontsize=18)
ax.legend(fontsize=13,ncol = 3, loc='upper left')
ax2.legend(fontsize=13, ncol = 3, loc="upper right")
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
plt.tight_layout()
plt.show()
fig.savefig('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/RESULTS/ottobre2024.png')



print("Done!\nYou can find the final data and graphs in the same folder as this script.")
```

```python
corrections = corrections.set_index(pd.to_datetime(corrections['datetime']))
corrections["neutrons_incoming"] = corrections["neutrons"]*corrections["incoming_correction"]
corrections["neutrons_RH"] = corrections["neutrons"]*corrections["RH_correction"]
corrections["neutrons_press"] = corrections["neutrons"]*corrections["press_correction"]
corrections['avg_neutrons_inc'] = corrections['neutrons_incoming'].rolling(12, center=True).mean()
corrections['avg_neutrons_RH'] = corrections['neutrons_RH'].rolling(12, center=True).mean()
corrections['avg_neutrons_press'] = corrections['neutrons_press'].rolling(12, center=True).mean()
corrections['sm_incoming'] = (A0/((corrections["avg_neutrons_inc"]/N0)-A1)-A2-theta_offset)
corrections['sm_RH'] = (A0/((corrections["avg_neutrons_RH"]/N0)-A1)-A2-theta_offset)
corrections['sm_press'] = (A0/((corrections["avg_neutrons_press"]/N0)-A1)-A2-theta_offset)

fig, ax = plt.subplots(figsize=(22, 5))
x = corrections.index

ax.plot(x, corrections['incoming_correction'], label='incoming', color ="red")
ax.plot(x, corrections['RH_correction'], label='RH', color ="green")
ax.plot(x, corrections['press_correction'], label='press', color ="blue")



ax.set_ylabel("soil moisture",fontsize=12)
ax.set_title("Confronto diverse correzioni",fontsize=18)
ax.legend(fontsize=13,ncol = 3, loc='upper left')
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
plt.tight_layout()
plt.show()
fig.savefig('C:/Users/Caterina/Desktop/neutroni_dati/SCRIPT/RESULTS/confronto_correzioni.png')
```
